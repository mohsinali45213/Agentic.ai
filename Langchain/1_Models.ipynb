```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(
  model="gemini-2.5-flash",
  temperature=0,  # 0.0 to 2.0 ; lower is more deterministic and higher is more creative
  max_output_tokens=1024,  # max tokens in the output
  
)

response = model.invoke("Write a 5 line poem about Earth.")
print(response.content)
```
